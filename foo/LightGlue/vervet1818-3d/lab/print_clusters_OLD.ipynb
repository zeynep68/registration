```python
#

path = "data/aa/volume/agglomerative_9/"
resolution = 0.0843776  # in mm
cmap = "tab10"

rename = False

##

from skimage.transform import rescale
import matplotlib.pyplot as plt
import numpy as np
import h5py as h5
import os

for d in os.listdir(path):
    p = os.path.join(path, d)
    for f in os.listdir(p):
        if f.endswith('h5'):

            old_f = os.path.join(p, f)
            model_name = p.split('/')[-1]
            if rename:
                new_f = os.path.join(p, model_name + "_" + f)
                print(old_f, new_f)
                os.rename(old_f, new_f)
            else:
                new_f = old_f
            os.makedirs(os.path.join(path, 'thumbnails'), exist_ok=True)
            with h5.File(new_f, 'r') as h5f:
                volume = h5f['volume']
                spacing = volume.attrs['spacing']
                scale = tuple(s / resolution for s in spacing)
                print(scale)
                shape = volume.shape
                mid0 = shape[0] // 2
                mid1 = shape[1] // 2
                mid2 = shape[2] // 2
                im0 = np.transpose(volume[:, :, mid2])[::-1, ::-1]
                im0[im0 == 15] = 18
                print(np.unique(im0))
                im0 = rescale(im0, scale=(scale[1], scale[0]), anti_aliasing=False, preserve_range=True, order=0).astype(np.uint8)
                print(np.unique(im0))
                plt.imsave(os.path.join(path, 'thumbnails', model_name + '_0.png'), im0, cmap=cmap, vmin=1, vmax=21)
                plt.imshow(im0, cmap=cmap, vmin=1, vmax=21, interpolation='nearest')
                plt.show()
                im1 = np.transpose(volume[mid0, :, :])[::-1, ::-1]
                im1[im1 == 15] = 18
                im1 = rescale(im1, scale=(scale[2], scale[1]), anti_aliasing=False, preserve_range=True, order=0).astype(np.uint8)
                plt.imsave(os.path.join(path, 'thumbnails', model_name + '_1.png'), im1, cmap=cmap, vmin=1, vmax=21)
                plt.imshow(im1, cmap=cmap, vmin=1, vmax=21, interpolation='nearest')
                plt.show()
                im2 = np.transpose(volume[:, mid1, :])[::-1, ::-1]
                im2[im2 == 15] = 18
                im2 = rescale(im2, scale=(scale[2], scale[0]), anti_aliasing=False, preserve_range=True, order=0).astype(np.uint8)
                plt.imsave(os.path.join(path, 'thumbnails', model_name + '_2.png'), im2, cmap=cmap, vmin=1, vmax=21)
                plot = plt.imshow(im2, cmap=cmap, vmin=1, vmax=21, interpolation='nearest')
                colorbar = plt.colorbar(plot, ticks=np.unique(im2)[1:], boundaries=np.arange(23) - 0.5)
                plt.savefig("legend.svg")
                plt.show()
```
